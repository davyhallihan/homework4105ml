```python
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/4105a0/D3.csv')
```

```python
X1 = df.values[:, 0]
X2 = df.values[:, 1]
X3 = df.values[:, 2]
truth = df.values[:, 3]
records = len(truth)

bias = np.ones((records, 1))
```

```python
def cost(explanatory, truth, theta):
    predictionlist = np.dot(explanatory, theta)
    serrorlist = np.square(np.subtract(predictionlist, truth))
    cost = 0.5 * np.sum(serrorlist) / records
    return cost
```

```python
def gradient_descent(explanatory, truth, theta, alpha, iterations, records):
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = np.dot(explanatory, theta)
        errors = np.subtract(predictions, truth)
        theta -= (alpha / records) * np.dot(explanatory.transpose(), errors)
        cost_history[i] = cost(explanatory, truth, theta)

    return theta, cost_history
```

```python
# Prepare Input Explanatory Variable 1
prep1 = X1.reshape(records, 1)
explanatory1 = np.hstack((bias, prep1))

# Output Variables
theta1 = np.zeros(2)

# Tuning Variables
iterations = 1500
alpha = 0.02

# Calculate initial cost
cost1 = cost(explanatory1, truth, theta1)
print('X1 Starting cost =', cost1)

# Train
theta1, cost_history1 = gradient_descent(explanatory1, truth, theta1, alpha, iterations, records)
print('X1 Final value of theta =', theta1)
print('X1 cost_history =', cost_history1)
```

```python
# X1 Plots

# X1 Fit Plot
plt.scatter(explanatory1[:, 1], truth, color='red', marker='+', label='Training Data') #Scatter Data
plt.plot(explanatory1[:, 1], explanatory1.dot(theta1), color='green', label='Linear Regression') #Model
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('X1 Linear Regression Fit (Problem 1, Part 2 Plot)')
plt.legend()

# Show the plot
plt.show()

# X1 Loss Plot
plt.plot(range(1, iterations + 1), cost_history1, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)

plt.xlabel('Number of iterations')
plt.ylabel('Cost')
plt.title('X1 Loss over Iterations (Problem 1, Part 2 Plot)')

# Show the plot
plt.show()
```
